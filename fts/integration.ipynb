
## Integration schemes
Fields are updated using one of three update schemes: explicit Euler (EM), first-order semi-implicit (1S), and a Euler predictor-corrector (EMPC) scheme. All of the schemes rely on a numerical update of the field $w(r)$ according to a fictitious time,

$$
\frac{\partial w(t)}{\partial t}
= 
-\frac{\delta \mathcal{H}}{\delta w},
$$

where the spatial dependence of $w(r)$ has been suppressed for brevity. The update schemes will be presented here without discussion or derivation, and we refer the interested reader to the monograph by Fredrickson and Delaney for further details.


### Euler update scheme (EM)
The Euler update scheme is the most straightforward. The time derivative in the field update equation above is approximated using an explicit discretization scheme, 

$$
\frac{w_{t+\delta t} - w_t } { \delta t}
= - \left(\frac{\delta \mathcal{H}}{\delta w} \right)_t.
$$

This equation is then explicitly solved for the fields at time $t+\delta t$ given the fields at time $t$,

$$
w_{t+\delta t} = 
w_t - \delta t \left(\frac{\delta \mathcal{H}}{\delta w} \right)_t.
$$

This scheme can be straightforwardly implemented whether the fields are in real or Fourier space.

## Creating new potentials

There are several steps to implementing new potentials that must be taken into account, and different details must be implemented (or their absense safely caught) depending on the integration schemes to be used. The minimum requirements are:
1. A header file containing function declarations, fts_potential_NAME.h, including those that need to be overriden from the parent class. This should be followed by the definition of any needed parameters. For example, the Helfand potentials requires the definition of $\kappa N_r$ and the size of the time step for the field associated with this potential.

2. A source file containing function definitions, fts_potential_NAME.cu. The constructor (PotentialNAME(...)) should handle parsing all of the input command. 

3. For the simple EM update scheme, one must include the update field routine, which in turn requires the creation of the CUDA device routine ``d_makeNAMEForce()''. The contribution of this term to the model Hamiltonian should also be implemented.

4. Advanced update schemes: If one plans to implement the 1S update scheme for this potential, the linear coefficients must be defined for each molecule type. See the Helfand and Flory potentials for examples of this. If one plans to implement the EMPC method, one must allocate memory for storage of the previous form of the potential field, its effective force, and (if compatible with complex Langevin) the random noise field.
